```python
group_multiple = df.groupby(by='CustomerID').agg({'Quantity':['mean'],'Price':['mean']})
group_multiple.columns = {'Price_mean','Quantity_mean'}
group_multiple  =  group_multiple.rename(columns={'Price_mean':'Quantity_mean','Quantity_mean':'Price_mean'})
group_multiple = group_multiple.reset_index()
```

```python
group_multiple.describe()
```

```python
Q1_Qm = group_multiple['Quantity_mean'].quantile(0.25)
Q3_Qm = group_multiple['Quantity_mean'].quantile(0.75)
IQR_m = Q3_Qm - Q1_Qm
```

```python
IQR_m
```

```python
lower_limit_Qm = Q1_Qm - 1.5 * IQR_m
upper_limit_Qm = Q3_Qm + 1.5 * IQR_m
print(lower_limit_Qm)
print(upper_limit_Qm)
```

```python
group_multiple = group_multiple[(group_multiple['Quantity_mean'] > lower_limit_Qm) & (group_multiple['Quantity_mean'] < upper_limit_Qm)]
```

```python
group_multiple.describe()
```

```python
Q1_Pm = group_multiple['Price_mean'].quantile(0.25)
Q3_Pm = group_multiple['Price_mean'].quantile(0.75)
IQR_Pm = Q3_Pm - Q1_Pm
lower_limit_Pm = Q1_Pm - 1.5 * IQR_Pm
upper_limit_Pm = Q3_Pm + 1.5 * IQR_Pm
group_multiple = group_multiple[(group_multiple['Price_mean'] > lower_limit_Pm) & (group_multiple['Price_mean'] < upper_limit_Pm)]
```

```python
group_multiple.describe()
```

```python
sns.boxplot(x=group_multiple['Quantity_mean'])
```

```python
sns.boxplot(x=group_multiple['Price_mean'])
```

```python
df = df[(df['Price'] > lower_limit_Pm) & (df['Price'] < upper_limit_Pm)]
df = df[(df['Quantity'] > lower_limit_Qm) & (df['Quantity'] < upper_limit_Qm)]
df=df.reset_index(drop=True)
```

```python
df_master_top10 = pd.DataFrame()
for i in df_master_plot.Id.unique():
    df_master_top10 = df_master_top10.append(df_master_plot[df_master_plot['Id']==i].head(10))
```

```python
df_master_top10 = pd.merge(df_master_top10,df_locations,left_on='Id',right_on='Id')
```

```python
df_master_top10
```

```python
for i in df_master_top10.Id.unique():
    df_plot_temp = df_master_top10[df_master_top10['Id']==i]
    df_plot_temp = df_plot_temp.reset_index(drop=True)
    name = str(df_plot_temp['Id'].unique())
    df_plot_temp.sort_values(by='total_revenue',ascending=True).plot(kind='barh',x='Itemname',y='total_revenue',title=name)
```
